### Univariate

```python
df.info()
```

```python
for i, predictor in enumerate(df[['state','area_code','international_plan','voice_mail_plan']]):
    plt.figure(i)
    plt.figure(figsize=(16,6))
    sns.countplot(data=df, x=predictor, hue='churn')
```

```python
df['churn']=np.where(df.churn=='yes',1,0)
df
```

```python
df_dummies = pd.get_dummies(df)
df_dummie
```

```python
df_dummies.info()
```

```python
for i in enumerate(df_dummies[['account_length','number_vmail_messages','total_day_minutes','total_day_calls','total_day_charge','total_eve_minutes','total_eve_calls','total_eve_charge','total_night_minutes','total_night_calls','total_night_charge','total_intl_minutes','total_intl_calls','total_intl_charge','number_customer_service_calls']]):
    plt.figure(figsize=(16,5))
    univariate_num = sns.kdeplot(df_dummies[i[1]][(df_dummies['churn']==0)], color='r',shade=True)
    univariate_num = sns.kdeplot(df_dummies[i[1]][(df_dummies['churn']==1)], ax=univariate_num, color='b',shade=True)
    univariate_num.legend(['No churn','Churn'], loc='upper right')
    univariate_num.set_ylabel('Density')
    univariate_num.set_xlabel('')
    univariate_num.set_title(i[1] + ' by Churn')
    plt.show()
```

```python
plt.figure(figsize=(26,10))
df_dummies.corr()['churn'].sort_values(ascending = False).plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(18,6))
df.corr()['churn'].sort_values(ascending = False).plot(kind='bar')
plt.show()
```
